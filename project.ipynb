```python
numeric_cols = df.select_dtypes(include=['float64', 'int64'])
df = df[(np.abs(stats.zscore(numeric_cols)) < 3).all(axis=1)]
```

## Analysing the dataset by filtering profession "Financial Analyst"

```python
filtered_finance = df.query("Job_Title == 'Financial Analyst'")
```

```python
plt.figure(figsize=(11,5))
plt.bar(avg_by_exp_finance['Years_Experience'], avg_by_exp_finance['Average_Salary'], color = 'turquoise')
plt.xlabel('Years of Experience')
plt.ylabel('Average Salary')
plt.title('Average Salary by Years of Experience')
plt.show()
```

```python
plt.figure(figsize=(11,5))
plt.bar(filtered_finance['Education_Level'], filtered_finance['Average_Salary'], color = 'purple')
plt.xlabel('Education Level')
plt.ylabel('Average Salary')
plt.title('Average Salary by Level of Educationfor Financial Analyst')
plt.show()
```

# Analysis with respect to "Job_Titles"

```python
jobs_by_salary = df.groupby(['Job_Title'])['Average_Salary'].mean().to_frame()
jobs_by_salary.sort_values(by='Average_Salary', inplace=True)
jobs_by_salary.reset_index(inplace=True)
```

```python
fig, ax = plt.subplots(figsize=(11,6))
x = jobs_by_salary['Job_Title']
y =  jobs_by_salary['Average_Salary']
ax.plot(x, y)
ax.set_title('Average Salary by Professions')
ax.set_xlabel('Job Titles')
ax.set_ylabel('Salary')
ax.set_xticks(jobs_by_salary['Job_Title'])
ax.set_xticklabels(labels = jobs_by_salary['Job_Title'], rotation = (45), fontsize = 8, ha='right')
plt.show()
```

```python
job_by_ai = df.groupby('Job_Title')['AI_Exposure_Index'].mean().to_frame()
job_by_ai.sort_values(by='AI_Exposure_Index', inplace=True)
job_by_ai.reset_index(inplace=True)
```

```python
fig, ax = plt.subplots(figsize=(11,6))
x = job_by_ai['Job_Title']
y = job_by_ai['AI_Exposure_Index']
ax.plot(x, y)
ax.set_title('AI Exposure Index')
ax.set_xlabel('Job Titles')
ax.set_ylabel('AI Exposure')
ax.set_xticks(job_by_ai['Job_Title'])
ax.set_xticklabels(labels = job_by_ai['Job_Title'], rotation = (45), fontsize = 8, ha='right')
plt.show()
```

```python
job_by_auto = df.groupby('Job_Title')['Automation_Probability_2030'].mean().to_frame()
job_by_auto.sort_values(by='Automation_Probability_2030', inplace=True)
job_by_auto.reset_index(inplace=True)
```

```python
fig, ax = plt.subplots(figsize=(11,6))
x = job_by_auto['Job_Title']
y = job_by_auto['Automation_Probability_2030']
ax.plot(x, y)
ax.set_title('Automation Probability in 2030')
ax.set_xlabel('Job Titles')
ax.set_ylabel('Automation Probability')
ax.set_xticks(job_by_auto['Job_Title'])
ax.set_xticklabels(labels = job_by_auto['Job_Title'], rotation = (45), fontsize = 8, ha='right')
plt.show()
```
